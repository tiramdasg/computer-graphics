## 4. Arrays

Often one has to compute large amounts of data. To do so one often uses arrays.

In Julia there are many ways to initialize arrays.

````julia
# initialization by value
a1 = [1,2,3,4,5,6.0]
println("a1 = $a1")
# initialization by type and value
a2 = Float32[1 2 3;4 5 6]
println("a2 = $a2")
# initialize an array of a certain type with all entries set to zero 
a3 = zeros(Int64,2,3) 
println("a3 = $a3")
# allocation by type 
a4 = Array{Float32,2}(undef,2,3)
println("a4 = $a4")
# construct arrays using comprehensions 
a5 = [r*c^2 for r=1:4,c=1:4]
println("a5 = $a5")
````

Depending on the initialization arrays can have different element types, though they represent the same data.

We can find out about the type of a variable using the function `typeof`.

````julia
println("The type of a1 is $(typeof(a1))")
println("The type of a2 is $(typeof(a2))")
println("The type of a3 is $(typeof(a3))")
println("The type of a4 is $(typeof(a4))")
println("The type of a5 is $(typeof(a5))")
````

To are several ways to access an array partially.

````julia
# Access the whole array
a5
````

````julia
# access a single row a the matrix a5
a5[:,1]
````

````julia
# access elements 1 to 3 of column 1
a5[1:end-1,1]
````

````julia
# accessing the array linearly
a5[1:8]
````

It is possible to perform element wise operations

````julia
println.(a1)
a1 .+= 1
println(a1)
````

or vector operations

````julia
# scalar multiplication of a2 by 1/2
println(a2)
a2 /= 2
println(a2)
# vector addition
a2 = reshape(a2,6)
println(a1+a2)
# many more possible
````

## 6. Functions

In Julia, a function is an object that maps a tuple of argument values to a return value. Julia functions are not pure mathematical functions, in the sense that functions can alter and be affected by the global state of the program. The basic syntax for defining functions in Julia is:

````julia
function minus(a,b)
    return a-b
end
# or using the short notation
plus(a,b) = a+b

println(plus(1,1))
println(minus(1,1))
````

Julia has a sophisticated type system and allows multiple dispatch on argument types.

````julia
function foo(x::Int128)
    println("The argument type is Int128.")
end

function foo(x::Int64)
    println("The argument type is Int64.")
end

function foo(x::Signed)
    println("The argument type is either of Int32, Int16, Int8.")
end

function foo(x)
    println("The argument type is not Signed.") 
end
````

````julia
x1 = zero(Int128)
x2 = 0
x3 = zero(Int8)
x4 = UInt32(1)

foo(x1)
foo(x2)
foo(x3)
foo(x4)
````

In many cases, function arguments have sensible default values and therefore might not need to be passed explicitly in every call.

````julia
function bar(x,y=0,z=0)
    return [x,y,z] 
end
````

This allows to call the function with 1, 2 or 3 arguments.

````julia
bar(1)
````

````julia
bar(1.0,1.0)
````

````julia
bar(1,1,1)
````

Some functions need a large number of arguments, or have a large number of behaviors. Remembering how to call such functions can be difficult. Keyword arguments can make these complex interfaces easier to use and extend by allowing arguments to be identified by name instead of only by position.

````julia
# functions with keyword arguments are defined 
# using a semicolon in the signature
function braket(x;bra="<",ket::String=">")
    println("$bra $x $ket") 
end
````

````julia
braket(1)
````

````julia
braket(0,ket="1",bra=1)
````

It is a convention in Julia that functions ending with `!` modify at least one of their arguments.

````julia
function set42!(a)
    a[1] = 42
end

A = [1,2,3]; println(A);
set42!(A); println(A);
````
